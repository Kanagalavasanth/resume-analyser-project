```python
# Install required libraries
!pip install PyPDF2
!pip install language_tool_python
!pip install rapidfuzz

from google.colab import files
from PyPDF2 import PdfReader
import language_tool_python
from rapidfuzz import fuzz

# Function to extract text from a PDF
def extract_text_from_pdf(file_name):
    try:
        reader = PdfReader(file_name)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
        return text
    except Exception as e:
        return f"Error extracting text from PDF: {e}"

# Function to generate feedback
def generate_feedback(resume_text):
    # Initialize the grammar correction tool
    tool = language_tool_python.LanguageTool("en-US")
    corrected_text = tool.correct(resume_text)

    # Example job description keywords
    job_keywords = ["Python", "Machine Learning", "Data Analysis", "AI", "NLP"]

    # Extract keywords from the resume text
    extracted_keywords = [word for word in job_keywords if word.lower() in resume_text.lower()]

    # Match extracted keywords with job description
    matched_keywords = list(set(extracted_keywords) & set(job_keywords))

    # Calculate ATS score based on keyword matching
    ats_score = len(matched_keywords) / len(job_keywords) * 100

    return {
        "Extracted Keywords": extracted_keywords,
        "Corrected Text": corrected_text,
        "Matched Keywords": matched_keywords,
        "ATS Score": round(ats_score, 2),
    }

# Upload the resume file
print("Upload your PDF resume:")
uploaded_file = files.upload()

if uploaded_file:
    file_name = list(uploaded_file.keys())[0]
    resume_text = extract_text_from_pdf(file_name)

    if resume_text.startswith("Error"):
        print(resume_text)  # Show error message if extraction fails
    else:
        feedback = generate_feedback(resume_text)

        # Display results
        print("\n--- Extracted Keywords ---")
        print(feedback["Extracted Keywords"])

        print("\n--- Corrected Resume Text ---")
        print(feedback["Corrected Text"])

        print("\n--- Matched Keywords ---")
        print(feedback["Matched Keywords"])

        print(f"\n--- ATS Score: {feedback['ATS Score']} / 100 ---")
```
